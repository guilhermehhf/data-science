##### Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans, DBSCAN
```

##### Dataset Graph

```python
df = pd.read_csv('paint_data.csv')

plt.scatter(df['x'],df['y'],color = 'blue',alpha = 0.7)
plt.show()
```

##### KMeans

```python
kmeans = KMeans(n_clusters=4)
kmeans.fit(df)

plt.scatter(df['x'],df['y'], c=kmeans.labels_, s=50, cmap='viridis',alpha=0.6)
centroides = kmeans.cluster_centers_
plt.scatter(centroides[:, 0], centroides[:, 1], c='black', s=200, alpha=0.7);
plt.show()
```

##### DBSCAN

```python
dbscan = DBSCAN(eps=0.04, min_samples=len(df.columns)*2).fit(df)
dbscan.labels_
plt.scatter(df['x'],df['y'], c=dbscan.labels_, s=50,alpha=0.7)
plt.show()
```
